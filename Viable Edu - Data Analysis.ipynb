# Construct Classification Targets

_The objective of this project is to classify eacn individual by experience level, industry, and affiliation. We can then categorize these groups by histograms in order to understand the relative categorical description of the userbase._

```python
experience_level = ['Student / Intern: 1', 'Entry-level: 2', 'Mid-level: 3', 'Senior-level: 4', 'Executive: 5']
```

```python
industry_type = ['Business: 1', 'Finance: 2', 'Consulting: 3', 'Engineering & Technology: 4', 'Government: 5', 
                 'Healthcare: 6', 'Education: 7', 'Legal: 8', 'Trades and Transportation: 9', 
                 'Protective Services: 10', 'Entertainment: 11', 'Academic / Research: 12', 
                 'Social Activism: 13', 'Other: 14', 'N/A: 15']
```

# Classify Followers by Type

```python
job = followers['jobTitle'].tolist()
```

```python
job
```

```python
def classify_jobs(job_list, experience, industry):
    experience_list = []
    industry_one = []
    industry_two = []
    affiliation_list = []
    for item in job_list:
        print(item)
        print('What is this individuals level of experience?')
        for obj in experience:
            print(obj)
        choice = int(input('Please select the number corresponding to your choice'))
        selection = experience[choice - 1]
        experience_list.append(selection)
        print('What two industries best match this users job title?')
        for obj in industry:
            print(obj)
        choice_one = int(input('Please select the number corresponding to your first choice'))
        choice_two = int(input('Please select the number corresponding to your second choice'))
        industry_one.append(industry[choice_one - 1])
        industry_two.append(industry[choice_two - 1])
        affiliation = input('What organization is this individual affiliated with? Enter N/A if not applicable')
        affiliation_list.append(affiliation)
    return experience_list, industry_one, industry_two, affiliation_list
```
